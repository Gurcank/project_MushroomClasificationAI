# Mushroom Toxicity Model Training

This notebook trains and compares models for binary mushroom toxicity classification.

Target mapping: `e = 0` (edible), `p = 1` (poisonous).
Primary selection criterion: minimize false negatives, then maximize recall and F1.

```python
import pandas as pd
from src.preprocessing import load_raw_dataset, preprocess_dataframe, split_dataset
from src.model import train_and_select_best

raw_df = load_raw_dataset('data/mushrooms.csv')
clean_df = preprocess_dataframe(raw_df)
bundle = split_dataset(clean_df)
bundle.X_train.head()
```

```python
result = train_and_select_best(
    X_train=bundle.X_train,
    y_train=bundle.y_train,
    X_test=bundle.X_test,
    y_test=bundle.y_test,
)
result.best_model_name
```
